```python
# sw en python
opcion= int(input("ingrese una opcion del 1 al 3"))
if opcion==1:
    print("opcion 1")
elif opcion==2:
    print("opcion 2")
elif opcion==3:
    print("opcion 3")
else:
    print("opcion no valida")
```

```python
# swiches con match en uso con python
valor=int(input("ingrese un valor del 1 al 3"))
match valor:
    case 1:
        print("opcion 1")
    case 2:
        print("opcion 2")
    case 3:
        print("opcion 3")
    case _:
        print("opcion no valida")
```

```python
# tenemos dos valores para sumar usando lambda
sumar=lambda x,y: x+y
sumar(2,3)
```

```python
# usando lambda verificar si es mayor o menor de edad
esMayor=lambda x: x>=18
esMayor(17)
```

```python
# Con lambda verificar si un numero ingresado es negativo, positivo o cero
positivo=lambda x: "positivo" if x>0 else "negativo" if x<0 else "cero"
print(positivo(-5))
print(positivo(0))
print(positivo(5))
```
